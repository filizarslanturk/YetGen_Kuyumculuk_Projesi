```python
import requests, time
from bs4 import BeautifulSoup
import datetime
from colorama import Fore, Style

# Şu anki tarih ve saat bilgisi
bugun = datetime.datetime.now()
# Tarih formatını ayarlayalım
tarih_formati = "{}.{}.{}".format(bugun.day, bugun.month, bugun.year)
# Saat formatını ayarlayalım
saat_formati = "{}:{}".format(bugun.hour, bugun.minute)
# Gram Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/gram-altin"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
gram_altin = soup.find("span", {"data-socket-attr" : "bid"})
gram_altin = gram_altin.text.strip().replace('.','').replace(',','.')
gram_altin=float(gram_altin)
# Çeyrek Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/ceyrek-altin"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
ceyrek_altin = soup.find("span", {"data-socket-attr" : "bid"})
ceyrek_altin = ceyrek_altin.text.strip().replace('.','').replace(',','.')
ceyrek_altin=float(ceyrek_altin)
# Yarım Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/yarim-altin"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
yarim_altin = soup.find("span", {"data-socket-attr" : "bid"})
yarim_altin = yarim_altin.text.strip().replace('.','').replace(',','.')
yarim_altin=float(yarim_altin)
# Tam Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/tam-altin"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
tam_altin = soup.find("span", {"data-socket-attr" : "bid"})
tam_altin = tam_altin.text.strip().replace('.','').replace(',','.')
tam_altin=float(tam_altin)
# Cumhuriyet Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/cumhuriyet-altini"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
cumhuriyet_altin = soup.find("span", {"data-socket-attr" : "bid"})
cumhuriyet_altin = cumhuriyet_altin.text.strip().replace('.','').replace(',','.')
cumhuriyet_altin=float(cumhuriyet_altin)
# 22 Ayar Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/22-ayar-bilezik"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
a_22_altin = soup.find("span", {"data-socket-attr" : "bid"})
a_22_altin = a_22_altin.text.strip().replace('.','').replace(',','.')
a_22_altin=float(a_22_altin)
# 18 Ayar Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/18-ayar-altin"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
a_18_altin = soup.find("span", {"data-socket-attr" : "bid"})
a_18_altin = a_18_altin.text.strip().replace('.','').replace(',','.')
a_18_altin=float(a_18_altin)
# 14 Ayar Altının fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/14-ayar-altin"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
a_14_altin = soup.find("span", {"data-socket-attr" : "bid"})
a_14_altin = a_14_altin.text.strip().replace('.','').replace(',','.')
a_14_altin=float(a_14_altin)

fiyatlar = {
    "Gram altın": gram_altin,
    "Çeyrek altın": ceyrek_altin,
    "Yarım altın": yarim_altin,
    "Tam altın": tam_altin, 
    "Cumhuriyet altını": cumhuriyet_altin,
    "22 ayar altın": a_22_altin,
    "18 ayar altın": a_18_altin,
    "14 ayar altın": a_14_altin
    } 
def display_fiyatlar():
    print(Fore.GREEN + Style.BRIGHT + "{} tarihi ve saat {} itibariyla güncel altın fiyatları:".format(tarih_formati, saat_formati) + Style.RESET_ALL)
    for item in list(fiyatlar.keys()):
        print(Fore.GREEN + Style.DIM + f" {item} : {fiyatlar[item]} TL" + Style.RESET_ALL)
 
while True:
    print("Lütfen yapmak için bir işlem seçin:")
    print("1. Güncel Altın Fiyatları\n2. Hazır Altın Ör: Gram,Yarım,Çeyrek...\n3. Ziynet Eşyası Ör: Kolye,Bilezik,Küpe,Yüzük\n4. Hurda-Hasarlı Altın\n5. Ana Menüye Dön\n")
    try:
        option = int(input("Seçiminiz: "))
        if option == 1: #güncel altın fiyatları
            while True:
                try:
                    display_fiyatlar()
                    print("\n")
                    break
                except:
                    print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)
                
        elif option == 2: ##Hazır altınlar: adet seçme işlemi ile gram, çeyrek, yarım, tam, cumhuriyet altını seçimi yapılacak
            print("\nElinde Hangi Altından Var?\n")
            print("1. Gram Altın\n2. Çeyrek Altın\n3. Yarım Altın\n4. Tam Altın\n5. Cumhuriyet Altını\n")
            try:
                hazir_altin_secim = int(input("Seçiminiz: "))
                if hazir_altin_secim == 1:
                    adet = int(input("Lütfen kaç adet gram altın satmak istediğinizi girin: "))
                    toplam_fiyat = adet * gram_altin
                    print(Fore.GREEN + Style.BRIGHT + f"{adet} Adet Gram Altın İçin Sana Biçtiğim Toplam Fiyat: {toplam_fiyat} TL " + Style.RESET_ALL)
                    break           
                elif hazir_altin_secim == 2:
                    adet = int(input("Lütfen kaç adet çeyrek altın satmak istediğinizi girin: "))
                    toplam_fiyat = adet * ceyrek_altin
                    print(Fore.GREEN + Style.BRIGHT + f"{adet} Adet Çeyrek Altın İçin Sana Biçtiğim Toplam Fiyat: {toplam_fiyat} TL " + Style.RESET_ALL)
                    break                
                elif hazir_altin_secim == 3:
                    adet = int(input("Lütfen kaç adet yarım altın satmak istediğinizi girin: "))
                    toplam_fiyat = adet * yarim_altin
                    print(Fore.GREEN + Style.BRIGHT + f"{adet} Adet Yarım Altın İçin Sana Biçtiğim Toplam Fiyat: {toplam_fiyat} TL " + Style.RESET_ALL)             
                    break               
                elif hazir_altin_secim == 4:
                    adet = int(input("Lütfen kaç adet tam altın satmak istediğinizi girin: "))
                    toplam_fiyat = adet * tam_altin
                    print(Fore.GREEN + Style.BRIGHT + f"{adet} Adet Tam Altın İçin Sana Biçtiğim Toplam Fiyat: {toplam_fiyat} TL " + Style.RESET_ALL)
                    break
                elif hazir_altin_secim == 5:
                    adet = int(input("Lütfen kaç adet cumhuriyet altını satmak istediğinizi girin: "))
                    toplam_fiyat = adet * cumhuriyet_altin
                    print(Fore.GREEN + Style.BRIGHT + f"{adet} Adet Cumhuriyet Altını İçin Sana Biçtiğim Toplam Fiyat: {toplam_fiyat} TL " + Style.RESET_ALL)
                    break
                ##else
            except ValueError:
                print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)             
           
        elif option == 3: #Altından yapılmış ziynet eşyasının ayarını, işçiliğini ve gramını seçip sonuca ulaştırır.
            altin_turleri = ["14 ayar", "18 ayar", "22 ayar"]
            iscilik_turleri = ["Kötü", "Orta", "İyi", "Çok iyi", "Mükemmel"]
            print("Aşağıdaki altın türlerinden birini seçiniz:")
            for index, altin in enumerate(altin_turleri, start=1):
                print(f"{index}. {altin}")    
            while True:
                try:
                    altin_secim = int(input("Seçiminizi yapınız: "))
                    if altin_secim not in range(1, len(altin_turleri) + 1):
                        print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)                        
                    else:
                        secilen_altin = altin_turleri[altin_secim - 1]
                        break
                except ValueError:
                    print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)                       
            while True:
                try:
                    gram = float(input("Eşyanın gram değerini giriniz: "))
                    break
                except ValueError:
                    print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)        
            print("\n")
            print("Aşağıdaki işçilik türlerinden birini seçiniz:")
            for index, iscilik in enumerate(iscilik_turleri, start=1):
                print(f"{index}. {iscilik}")
            while True:
                altin_fiyat = {
                "14 ayar": a_14_altin,
                "18 ayar": a_18_altin,
                "22 ayar": a_22_altin
                }  
                iscilik_oran = {
                    "Kötü": 0.2,
                    "Orta": 0.5,
                    "İyi": 1,
                    "Çok iyi": 1.9,
                    "Mükemmel": 2.5
                }
                try:
                    iscilik_secim = int(input("Seçiminizi yapınız: "))
                    if iscilik_secim not in range(1, len(iscilik_turleri) + 1):
                        print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)                       
                    else:
                        secilen_iscilik = iscilik_turleri[iscilik_secim - 1]
                        break
                except ValueError:
                    print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)
            try:
                fiyat = (float(altin_fiyat[secilen_altin]) * gram) * (1 + iscilik_oran[secilen_iscilik])
                print("\n")
                print(Fore.GREEN + Style.BRIGHT + f"Bu ziynet eşyası için sana biçtiğim fiyat: {fiyat:.2f} TL " + Style.RESET_ALL)
                break
            except KeyError:
                print(Fore.RED + Style.DIM + "Bir hata oluştu. Lütfen tekrar deneyiniz." + Style.RESET_ALL)     
              
        elif option == 4:  #Hurda Altından gram fiyatına göre %20 aşağı satılır.
            print("Lütfen satılacak ziynet eşyasının ayrısını giriniz: ")
            print("1- 14 Ayar")
            print("2- 18 Ayar")
            print("3- 22 Ayar")
            print("4- 24 Ayar")
            print("\n")
            try:
                altin_ayari_secim = int(input("Seçiminiz: "))
                gram = float(input("Lütfen satılacak eşyanın gram miktarını giriniz: "))
                if altin_ayari_secim == 1:
                    fiyat = a_14_altin * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda altın için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL) 
                    break
                elif altin_ayari_secim == 2:
                    fiyat = a_18_altin * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda altın için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL) 
                    break
                elif altin_ayari_secim == 3:
                    fiyat = a_22_altin * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda altın için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL) 
                    break
                elif altin_ayari_secim == 4:
                    fiyat = gram_altin * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda altın için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)                    
                    break
                else:
                    print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)
            except ValueError:
                print(Fore.RED + Style.DIM + "Hatalı giriş yaptınız! Lütfen sayısal değerler girin." + Style.RESET_ALL)                             
                
        elif option == 5: #Programı kapatır.
            print(".·.Program kapatılıyor.·.")
            time.sleep(2)
            print("Program kapandı.") 
            break
        else:
            print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)
    except ValueError:
        print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)
```
