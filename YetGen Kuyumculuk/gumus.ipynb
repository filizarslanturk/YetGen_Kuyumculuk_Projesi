```python
import requests, time
from bs4 import BeautifulSoup
import datetime
from colorama import Fore, Style
# Şu anki tarih ve saat bilgisi
bugun = datetime.datetime.now()
# Tarih formatını ayarlayalım
tarih_formati = "{}.{}.{}".format(bugun.day, bugun.month, bugun.year)
# Saat formatını ayarlayalım
saat_formati = "{}:{}".format(bugun.hour, bugun.minute)
# 925 ayar Gümüşün fiyatını İnternet Sitesinden Çekiyoruz.
url = "https://altin.doviz.com/gumus"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
gram_gümüs = soup.find("span", {"data-socket-attr" : "bid"})
gram_gümüs = gram_gümüs.text.strip().replace('.','').replace(',','.')
gram_gümüs = float(gram_gümüs)
külce_gümüs = gram_gümüs + (gram_gümüs * (17/100))
sikke_gram = gram_gümüs + (gram_gümüs * (60/100))
a_gümüs_925 = gram_gümüs * (92.5/100)
a_gümüs_900 = gram_gümüs * (90/100)
a_gümüs_800 = gram_gümüs * (80/100)

fiyatlar = {
    "Gram Gümüş": gram_gümüs,
    "Külçe Gümüş": külce_gümüs,
    "Sikke Gümüş(işçiliksiz)": sikke_gram,
    "925 Ayar Gümüş": a_gümüs_925, 
    "900 Ayar Gümüş": a_gümüs_900,
    "800 Ayar Gümüş": a_gümüs_800
    } 
def display_fiyatlar():
    print(Fore.GREEN + Style.BRIGHT + "{} tarihi ve saat {} itibariyla güncel gümüş fiyatları:".format(tarih_formati, saat_formati) + Style.RESET_ALL)
    for item in list(fiyatlar.keys()):
        print(Fore.GREEN + Style.DIM + f" {item} : {fiyatlar[item]} TL " + Style.RESET_ALL)

while True:
    print("Lütfen yapmak için bir işlem seçin:")
    print("1. Güncel Gümüş Fiyatları\n2. Hazır Gümüş: Külçe ve Sikke\n3. Ziynet Eşyası Ör: Kolye,Bilezik,Küpe,Yüzük\n4. Hurda-Hasarlı Gümüş\n5. Çıkış\n")
    try:
        option = int(input("Seçiminiz: "))
        if option == 1: #güncel altın fiyatları
            while True:
                try:
                    display_fiyatlar()
                    print("\n")
                    break
                except:
                    print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)

        elif option == 2:           
            try: 
                print("\nElinde Hangi Gümüşten Var?")
                print("1. Külçe Gümüş\n2. Sikke Gümüş\n")
                hazir_gümüs_secim = int(input("Seçiminiz: "))
                if hazir_gümüs_secim == 1:
                    adet = int(input("Lütfen kaç adet külçe gümüş satmak istediğinizi girin: "))
                    gram = float(input("Lütfen satılacak külçenin gram miktarını giriniz: "))
                    toplam_fiyat = adet * gram * külce_gümüs          
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gramdan {adet} adet külçe gümüş için sana biçtiğim toplam fiyat: {toplam_fiyat} TL " + Style.RESET_ALL)
                    break        
                elif hazir_gümüs_secim == 2:
                    print("\nLütfen satılacak sikke gümüşün işçilik seviyesini seçin:")
                    print("1- İşçiliksiz")
                    print("2- Kötü işçilik")
                    print("3- Orta işçilik")
                    print("4- İyi işçilik")
                    print("\n")
                    sikke_ayari_secim = int(input("Seçiminiz: "))
                    adet = int(input("Lütfen kaç adet sikke gümüş satmak istediğinizi girin: "))
                    gram = float(input("Lütfen satılacak sikkenin gram miktarını giriniz: "))
                    if sikke_ayari_secim == 1:
                        fiyat = adet * sikke_gram * gram
                        print(Fore.GREEN + Style.BRIGHT + f"{gram} gramdan {adet} adet işçiliksiz sikke gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                        print()
                        break
                    elif sikke_ayari_secim == 2:
                        fiyat = adet * gram_gümüs * gram * 2
                        print(Fore.GREEN + Style.BRIGHT + f"{gram} gramdan {adet} adet kötü işçilikli sikke gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                        break
                    elif sikke_ayari_secim == 3:
                        fiyat = adet * gram_gümüs * gram * 3
                        print(Fore.GREEN + Style.BRIGHT + f"{gram} gramdan {adet} adet orta işçilikli sikke gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                        break
                    elif sikke_ayari_secim == 4:
                        fiyat = adet * gram_gümüs * gram * 3.5
                        print(Fore.GREEN + Style.BRIGHT + f"{gram} gramdan {adet} adet iyi işçilikli sikke gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                        break
                    else:
                        print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)  
            except KeyError:
                print(Fore.RED + Style.DIM + "Bir hata oluştu. Lütfen tekrar deneyiniz." + Style.RESET_ALL)       
                
        elif option == 3: #Gümüşten yapılmış ziynet eşyasının ayarını, işçiliğini ve gramını seçip sonuca ulaştırır. 
            gümüs_turleri = ["925 ayar", "900 ayar", "800 ayar"]
            iscilik_turleri = ["Kötü", "Orta", "İyi", "Çok iyi", "Mükemmel"]
            print("Elinizdeki ziynet eşyasının ayarını seçin:")
            for index, gümüs in enumerate(gümüs_turleri, start=1):
                print(f"{index}. {gümüs}")
            while True:
                try:
                    gümüs_secim = int(input("Seçiminizi yapınız: "))
                    if gümüs_secim not in range(1, len(gümüs_turleri) + 1):
                        print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)
                    else:                 
                        secilen_gümüs = gümüs_turleri[gümüs_secim - 1]
                        break
                except ValueError:
                    print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)
            while True:
                try:
                    gram = float(input("Eşyanın gram değerini giriniz: "))
                    break
                except ValueError:
                    print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)
            print("\n")
            print("Aşağıdaki işçilik türlerinden birini seçiniz:")
            for index, iscilik in enumerate(iscilik_turleri, start=1):
                print(f"{index}. {iscilik}")    
            while True:
                gümüs_fiyat = {
                    "925 ayar": a_gümüs_925,
                    "900 ayar": a_gümüs_900,
                    "800 ayar": a_gümüs_800
                }
                iscilik_oran = {
                    "Kötü": 0.5,
                    "Orta": 1,
                    "İyi": 3,
                    "Çok iyi": 4,
                    "Mükemmel": 5
                }
                try:
                    iscilik_secim = int(input("Seçiminizi yapınız: "))
                    if iscilik_secim not in range(1, len(iscilik_turleri) + 1):
                        print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)
                    else:
                        secilen_iscilik = iscilik_turleri[iscilik_secim - 1]
                        break
                except ValueError:
                    print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)         
            try:
                fiyat = (float(gümüs_fiyat[secilen_gümüs]) * gram) * (1 + iscilik_oran[secilen_iscilik])
                print("\n")
                print(Fore.GREEN + Style.BRIGHT + f"Bu ziynet eşyası için sana biçtiğim fiyat: {fiyat:.2f} TL " + Style.RESET_ALL)
                break
            except KeyError:
                print(Fore.RED + Style.DIM + "Bir hata oluştu. Lütfen tekrar deneyiniz." + Style.RESET_ALL)

        elif option == 4: #Hurda Gümüşler gram fiyatına göre %20 aşağı satılır.
            print("\nLütfen satılacak hasarlı gümüşün ayarını giriniz:")
            print("1- 1000 Ayar")
            print("2- 925 Ayar")
            print("3- 900 Ayar")
            print("4- 800 Ayar")
            print("\n")
            try:
                gümüs_ayari_secim = int(input("Seçiminiz: "))
                gram = float(input("Lütfen satılacak eşyanın gram miktarını giriniz: "))
                if gümüs_ayari_secim == 1:
                    fiyat = gram_gümüs * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                    break
                elif gümüs_ayari_secim == 2:
                    fiyat = a_gümüs_925 * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                    break
                elif gümüs_ayari_secim == 3:
                    fiyat = a_gümüs_900 * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                    break
                elif gümüs_ayari_secim == 4:
                    fiyat = a_gümüs_800 * gram * 0.8
                    print(Fore.GREEN + Style.BRIGHT + f"{gram} gram hurda gümüş için sana biçtiğim toplam fiyat: {fiyat} TL " + Style.RESET_ALL)
                    break    
                else:
                    print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)           
            except ValueError:
                print(Fore.RED + Style.DIM + "Hatalı giriş yaptınız! Lütfen sayısal değerler girin." + Style.RESET_ALL)

        elif option == 5: #Programı kapatır.
                print("...Ana Menüye Dönülüyor...")
                time.sleep(2)          
                break
        else:
            print(Fore.RED + Style.DIM + "Geçersiz bir seçim yaptınız!" + Style.RESET_ALL)
    except ValueError:
        print(Fore.RED + Style.DIM + "Geçerli olmayan giriş yaptınız! Lütfen bir sayı giriniz." + Style.RESET_ALL)
```
